```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
df['average score'] = df['total score']/3
df.head()
```

```python
reading_full = df[df['reading score'] == 100].shape[0]
writing_full = df[df['writing score'] == 100].shape[0]
math_full = df[df['math score'] == 100].shape[0]

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = df[df['reading score'] <= 20].shape[0]
writing_less_20 = df[df['writing score'] <= 20].shape[0]
math_less_20 = df[df['math score'] <= 20].shape[0]

print(f'Number of students with less than or equal to 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than or equal to 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than or equal to 20 marks in Reading: {reading_less_20}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
plt.subplot(131)
sns.histplot(data=df, x='math score', bins=30, kde=True, color='g')
plt.title("Math Score Distribution")
plt.subplot(132)
sns.histplot(data=df, x='reading score', bins=30, kde=True, color='b')
plt.title("Reading Score Distribution")
plt.subplot(133)
sns.histplot(data=df, x='writing score', bins=30, kde=True, color='r')
plt.title("Writing Score Distribution")
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))
plt.subplot(121)
sns.histplot(data=df,x='total score',bins=30,kde=True,color='y')
plt.subplot(122)
sns.histplot(data=df,x='total score',kde=True,hue='gender')
plt.show()
```

```python
import matplotlib.pyplot as plt
values = df['parental level of education'].value_counts().values
labels = df['parental level of education'].value_counts().index
plt.figure(figsize=(10, 6))
explode = [0, 0.3, 0, 0, 0, 0.4] 
colors = ['red', 'green', 'blue', 'yellow', 'purple', 'orange'] 
plt.pie(values, labels=labels, autopct="%.2f%%", normalize=True, shadow=True, explode=explode, colors=colors)
legend = plt.legend(labels, loc='upper right')
legend.set_bbox_to_anchor((1.3, 1.3))
plt.title("Parental Level of Education")
plt.show()
```
